### **Central Limit Theorem**

- 모집단에서 샘플을 여러번 뽑을때 모집단의 크기 $N$이 충분히 크다면 모집단의 분포와 상관없이 표본평균의 분포는 정규분포에 근사한다

수식으로 표현하면, 모집단에서 랜덤 변수 $X_1, X_2, ..., X_n$을 독립적으로 샘플링했을 때, 표본평균:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

$n \to \infty$일 때 다음의 정규분포에 수렴한다.

$$
\bar{X} \sim \mathcal{N} \left( \mu, \frac{\sigma^2}{n} \right)
$$

여기서:
- $\mu$ = 모집단의 평균  
- $\sigma^2$ = 모집단의 분산  

즉, 표본 크기가 충분히 크다면 **모집단의 원래 분포가 정규분포가 아니더라도** 표본평균의 분포는 정규분포에 가까워진다는 것이 CLT의 핵심

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(1000)

x_mean = []


n = 10000

for i in range(n):
    sample = np.random.choice(x, 10)
    
    x_mean.append(np.mean(sample))
    
    
plt.hist(x_mean, bins=30)
plt.show()
```
